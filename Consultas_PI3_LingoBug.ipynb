```python
import sqlite3
from sqlite3 import Error

# Função de alteração (insert, update, delete...)
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print(f"Query {query.split('(')[0].strip()} executada.")
        if cursor.rowcount != -1:
            print(f"{cursor.rowcount} linhas afetadas.")
    except Error as e:
        print(f"Erro: '{e}'")

# Função de leitura de dados (select)
def execute_read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as e:
        print(f"Erro: '{e}'")

connection = sqlite3.connect("lingobug.sqlite")

query = """
create table usuario(
    id_usuario integer primary key autoincrement,
    nome varchar(100),
    email varchar(100),
    senha varchar(100),
    foto varchar(255)
);
"""
execute_query(connection, query)

query = """
create table questao(
    id_questao integer primary key autoincrement,
    enunciado text,
    verdadeira varchar(255),
    falsa1 varchar(255),
    falsa2 varchar(255),
    falsa3 varchar(255)
);
"""
execute_query(connection, query)

query = """
create table responde(
    id_usuario integer,
    id_questao integer,
    resultado varchar(50),
    primary key (id_usuario, id_questao),
    foreign key (id_usuario) references usuario(id_usuario),
    foreign key (id_questao) references questao(id_questao)
);
"""
execute_query(connection, query)
```

Como ela será reutilizada em diferentes funcionalidades do sistema, uma View facilita a manutenção e evita repetir a mesma consulta SQL.**
* **Descrição da ação esperada ao rodar a consulta: A View retorna o nome do usuário, a questão respondida e o resultado obtido em cada resposta registrada.**

```python
## ESCREVA ABAIXO A VIEW
query = """
CREATE VIEW vw_historico_respostas AS
SELECT
    usuario.nome,
    questao.enunciado,
    responde.resultado
FROM usuario
INNER JOIN responde
    ON usuario.id_usuario = responde.id_usuario
INNER JOIN questao
    ON questao.id_questao = responde.id_questao;
"""
execute_query(connection, query)

## ESCREVA ABAIXO O SELECT QUE RETORNA TODOS OS ELEMENTOS DA VIEW
query = """
SELECT * FROM vw_historico_respostas;
"""

resultado = execute_read_query(connection, query)
print(resultado)
```

Crie um procedure relevante para sua aplicação em PI3. Descreva a ação esperada ao executar a procedure e, ao justificar o contexto, explique por que essa ação é tão recorrente no seu PI3 a ponto de precisar de uma procedure.

**-- CÉLULA EDITADA--**

* **Justificativa para que a operação seja um Procedure (Contexto de PI3): O registro das respostas dos usuários é uma operação realizada sempre que um teste de nivelamento é concluído. Utilizar uma procedure centraliza essa lógica, facilitando a manutenção e evitando repetição de código no sistema**

* **Descrição da ação esperada ao executar o procedure: A procedure recebe o usuário, a questão e o resultado obtido, armazenando essas informações na tabela responde.**

```python
## ESCREVA ABAIXO O PROCEDURE
CREATE PROCEDURE RegistroResposta(
    IN p_id_usuario INT,
    IN p_id_questao INT,
    IN p_resultado INT
)
BEGIN
    INSERT INTO responde(id_usuario, id_questao, resultado)
    VALUES(p_id_usuario, p_id_questao, p_resultado);
END;

## ESCREVA ABAIXO UMA EXECUÇÃO DA PROCEDURE
CALL RegistroResposta(1, 2, 1);
```

Crie uma function relevante para sua aplicação em PI3.
Descreva o valor esperado como retorno ao executar a function e, ao justificar o contexto, explique por que é importante ter essa lógica encapsulada para uso recorrente no seu PI3.

**-- EDITE ESSA CÉLULA --**

* **Justificativa para que a operação seja uma Function (Contexto de PI3): A pontuação obtida por um usuário é consultada diversas vezes para apresentar seu desempenho. Utilizar uma function permite reutilizar essa lógica sempre que necessário.**

* **Descrição da operação realizada e valor retornado pela Function: A function recebe o ID do usuário e retorna a quantidade total de respostas corretas registradas.**

```python
## ESCREVA ABAIXO A FUNCTION
CREATE FUNCTION Total_acertos(idUs INT)
RETURNS INT
BEGIN
    DECLARE total INT;

    SELECT COUNT(*)
    INTO total
    FROM responde
    WHERE id_usuario = idUs
    AND resultado = 1;

    RETURN total;
END;

## ESCREVA ABAIXO UMA CONSULTA QUE INCLUI A FUNCTION
SELECT
    nome,
    Total_acertos(id_usuario) AS acertos
FROM usuario;
```

Crie um trigger relevante para sua aplicação em PI3.
Descreva o evento que aciona o trigger e, ao justificar o contexto, explique por que essa automação é necessária para manter a integridade ou eficiência do seu PI3.

**-- EDITE ESSA CÉLULA --**

* **Justificativa para que a operação seja um Trigger (Contexto de PI3): No LingoBug, cada usuário deve responder uma questão apenas uma vez durante o teste de nivelamento. O trigger garante essa regra automaticamente, preservando a integridade dos dados e evitando registros duplicados.**

* **Descrição do evento que dispara o Trigger e das operações que ele executa: O trigger é acionado antes da inserção de um novo registro na tabela responde. Caso o usuário já tenha respondido aquela questão, a operação é cancelada e uma mensagem de erro é exibida.**

```python
## ESCREVA ABAIXO O TRIGGER
query = """
CREATE TRIGGER impedir_resposta_duplicada
BEFORE INSERT ON responde
FOR EACH ROW
WHEN EXISTS (
    SELECT 1
    FROM responde
    WHERE id_usuario = NEW.id_usuario
      AND id_questao = NEW.id_questao
)
BEGIN
    SELECT RAISE(ABORT, 'O usuário já respondeu essa questão.');
END;
"""
execute_query(connection, query)

## ESCREVA ABAIXO UM COMANDO SQL QUE DISPARA O TRIGGER
query = """
INSERT INTO responde(id_usuario, id_questao, resultado)
VALUES (1, 1, 1);
"""
execute_query(connection, query)
```
